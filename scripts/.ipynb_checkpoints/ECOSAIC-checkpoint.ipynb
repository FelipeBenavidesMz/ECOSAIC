# **ECOSAIC - Ecological Compression via Orthogonal Specialized Autoencoders for Interpretable Classification**

# 1. ECOSAIC model

```python
# =====================================================================
# OPTIMIZED DUAL AUTOENCODER: PURITY-FIRST + SOFT ORTHOGONALITY
#
# MAIN ROLE OF THIS SCRIPT:
# This autoencoder implements a dual-objective strategy that prioritizes
# the purity of LE1 (Biogeochemical) and LE2 (Hydrogeomorphological)
# while maintaining a minimum acceptable orthogonality between them.
#
# CATEGORY DEFINITIONS:
#   LE1 → BIOGEOCHEMICAL variables (physicochemical + biological)
#   LE2 → HYDROGEOMORPHOLOGICAL variables (hydrological + geomorphological)
# =====================================================================

import os
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import QuantileTransformer, RobustScaler, StandardScaler
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (
    Input, Dense, concatenate, Lambda, Dropout, BatchNormalization, Multiply, Subtract
)
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import l1_l2
import warnings
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, Callback, ModelCheckpoint
import logging
from datetime import datetime
import json
from sklearn.model_selection import train_test_split
import keras_tuner as kt
import gc
import shutil

# =====================================================================
# SECTION A: INITIAL CONFIGURATION AND ENVIRONMENT SETUP
# =====================================================================

GLOBAL_SEED = 42

def clean_tuner_directory():
    tuner_path = "data/checkpoints/tuner_optimized_dual"
    if os.path.exists(tuner_path):
        shutil.rmtree(tuner_path)
        print("Tuner directory cleaned")

clean_tuner_directory()

tf.keras.backend.clear_session()

try:
    physical_devices = tf.config.experimental.list_physical_devices('GPU')
    if physical_devices:
        for device in physical_devices:
            tf.config.experimental.set_memory_growth(device, True)
        print(f"GPU configured: {len(physical_devices)} device(s)")
    else:
        print("No GPUs found, using CPU")
except Exception as e:
    print(f"Error configuring GPU: {e}")

def set_global_seeds(seed=GLOBAL_SEED):
    import random
    os.environ['PYTHONHASHSEED'] = str(seed)
    random.seed(seed)
    np.random.seed(seed)
    tf.random.set_seed(seed)

    try:
        import rpy2.robjects as ro
        ro.r(f'set.seed({seed})')
    except Exception as e:
        print(f"Could not set R seed: {e}")

    print(f"Global seeds set: {seed}")

set_global_seeds(GLOBAL_SEED)

# =====================================================================
# SECTION B: RPY2 AND MINERVA CONFIGURATION FOR MIC CALCULATIONS
# =====================================================================

try:
    import rpy2
    import rpy2.robjects as ro
    from rpy2.robjects import pandas2ri
    from rpy2.robjects.packages import importr
    print("rpy2 available")

    pandas2ri.activate()
    ro.r('options(repos = c(CRAN = "https://cran.r-project.org"))')
    ro.r('options(warn = -1)')
    utils = importr('utils')

    try:
        if not ro.r('any(grepl("minerva", installed.packages()))')[0]:
            print("Installing minerva...")
            ro.r('install.packages("minerva", dependencies=TRUE, quiet=TRUE)')
        minerva = importr('minerva')
        print("Minerva R available")
        RPY2_AVAILABLE = True
    except Exception as e:
        print(f"Minerva error: {e}")
        RPY2_AVAILABLE = False
except ImportError:
    print("rpy2 not available - will use correlation as fallback")
    RPY2_AVAILABLE = False

# =====================================================================
# CATEGORY ASSIGNMENT:
#   LE1 (BIOGEOCHEMICAL)        → physicochemical_vars + biological_vars
#   LE2 (HYDROGEOMORPHOLOGICAL) → hydrogeomorphological_vars
# =====================================================================

# ------------------------------------------------------------------
# LE1: BIOGEOCHEMICAL VARIABLES
#   Physicochemical sub-group
# ------------------------------------------------------------------
physicochemical_vars = [
    'Phosphate', 'Iron', 'Light_bottom', 'Nitrate',
    'Dissolved_oxygen', 'pH', 'Salinity', 'Silicate', 'Temperature', 'P_a_r', 'Organic_matter'
]

# ------------------------------------------------------------------
# LE2: HYDROGEOMORPHOLOGICAL VARIABLES
# ------------------------------------------------------------------
hydrogeomorphological_vars = [
    'Aspect', 'Eastness', 'River_influence', 'Currents_velocity', 'Runoff',
    'Wind_v', 'Land_distance', 'L_t_m', 'Northness', 'R_d_m_v', 'Slope',
    'Stde', 'Superficie', 'Hardness', 'Particle_size', 'E_l_d', 'D_a_c',
    'R_i_m_b'
]

# ------------------------------------------------------------------
# LE1: BIOGEOCHEMICAL — Biological sub-group
# ------------------------------------------------------------------
biological_vars = [
    'Chlorophyll', 'Phytoplankton', 'Utilized_oxygen', 'Primary_productivity'
]

# Combined group assignments
biogeochemical_vars = physicochemical_vars + biological_vars   # → LE1
hydrogeo_vars       = hydrogeomorphological_vars               # → LE2

# Data file path
excel_file_path = "data/inputs/seafloor_environmental_variables.xlsx"

def setup_logging():
    logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
    return logging.getLogger(__name__)

logger = setup_logging()
warnings.filterwarnings("ignore")

print("VARIABLE CONFIGURATION:")
print(f"  LE1 — Biogeochemical        : {len(biogeochemical_vars)} variables")
print(f"  LE2 — Hydrogeomorphological : {len(hydrogeo_vars)} variables")
print(f"  Data file: {excel_file_path}")

# =====================================================================
# SECTION D: MIC FUNCTIONS FOR PURITY EVALUATION
# =====================================================================

def setup_signed_realistic_mic():
    if not RPY2_AVAILABLE:
        logger.warning("rpy2/minerva not available - will use correlation")
        return False
    try:
        r_mic_code = """
        suppressWarnings(library(minerva, quietly = TRUE))
        calculate_signed_realistic_mic <- function(original_df, latent_df) {
            orig_cols <- names(original_df)
            latent_cols <- names(latent_df)
            mic_matrix <- matrix(0.0, nrow = length(latent_cols), ncol = length(orig_cols))
            colnames(mic_matrix) <- orig_cols
            rownames(mic_matrix) <- latent_cols
            for (i in seq_along(latent_cols)) {
                y <- latent_df[[latent_cols[i]]]
                for (j in seq_along(orig_cols)) {
                    x <- original_df[[orig_cols[j]]]
                    if (any(is.na(x)) || any(is.infinite(x)) || sd(x) == 0 ||
                        any(is.na(y)) || any(is.infinite(y)) || sd(y) == 0) {
                        next
                    }
                    suppressWarnings({
                        tryCatch({
                            mine_result <- minerva::mine(x, y, alpha=0.6, C=15)
                            mic_value <- mine_result$MIC
                            correlation <- cor(x, y, method = "spearman")
                            signed_mic <- ifelse(correlation >= 0, mic_value, -mic_value)
                            mic_matrix[i, j] <- signed_mic
                        }, error = function(e) {})
                    })
                }
            }
            return(as.data.frame(mic_matrix))
        }
        """
        with open("/tmp/signed_realistic_mic.R", "w") as f:
            f.write(r_mic_code)
        ro.r.source("/tmp/signed_realistic_mic.R")
        logger.info("Signed MIC configured correctly")
        return True
    except Exception as e:
        logger.error(f"Error configuring signed MIC: {e}")
        return False

def calculate_signed_realistic_mic_matrix(latent_df, original_df):
    try:
        logger.info("Calculating signed MIC matrix...")
        original_clean = original_df.copy()
        latent_clean = latent_df.copy()

        for col in original_clean.columns:
            data = original_clean[col].values
            if np.var(data) > 1e-12:
                qt = QuantileTransformer(
                    n_quantiles=min(300, len(data)//2),
                    output_distribution='uniform',
                    random_state=GLOBAL_SEED
                )
                transformed = qt.fit_transform(data.reshape(-1, 1)).flatten()
                original_clean[col] = transformed

        for col in latent_clean.columns:
            data = latent_clean[col].values
            if np.var(data) > 1e-12:
                qt = QuantileTransformer(
                    n_quantiles=min(300, len(data)//2),
                    output_distribution='uniform',
                    random_state=GLOBAL_SEED
                )
                transformed = qt.fit_transform(data.reshape(-1, 1)).flatten()
                latent_clean[col] = transformed

        original_clean = original_clean.fillna(0.5).replace([np.inf, -np.inf], 0.5)
        latent_clean = latent_clean.fillna(0.5).replace([np.inf, -np.inf], 0.5)

        mic_result = ro.r['calculate_signed_realistic_mic'](original_clean, latent_clean)
        mic_matrix = pandas2ri.rpy2py(mic_result)

        logger.info(f"MIC matrix calculated: {mic_matrix.shape}")
        return mic_matrix
    except Exception as e:
        logger.error(f"Error calculating MIC: {e}")
        return calculate_signed_robust_correlation(latent_df, original_df)

def calculate_signed_robust_correlation(latent_df, original_df):
    logger.info("Calculating signed correlation...")
    corr_matrix = pd.DataFrame(index=latent_df.columns, columns=original_df.columns, dtype=float)

    for latent_col in latent_df.columns:
        for orig_col in original_df.columns:
            try:
                x = latent_df[latent_col].values
                y = original_df[orig_col].values
                from scipy.stats import spearmanr
                corr, _ = spearmanr(x, y)
                corr_matrix.loc[latent_col, orig_col] = corr if not np.isnan(corr) else 0.0
            except:
                corr_matrix.loc[latent_col, orig_col] = 0.0
    return corr_matrix

# =====================================================================
# SECTION E: DATA PROCESSOR WITH ANTI-CONTAMINATION TRANSFORMATIONS
# =====================================================================

class OptimizedDualProcessor:

    def __init__(self, excel_path):
        self.excel_path = excel_path
        self.df = None
        self.coordinates = None

    def load_excel_data(self):
        try:
            logger.info(f"Loading data: {self.excel_path}")
            self.df = pd.read_excel(self.excel_path)
            coord_cols = self.df.columns[:2]
            self.coordinates = self.df[coord_cols].copy()
            self.coordinates.columns = ['longitude', 'latitude']
            logger.info(f"Data loaded: {self.df.shape}")
            return True
        except Exception as e:
            logger.error(f"Error loading data: {e}")
            return False

    def anti_contamination_biogeochem_transform(self, data_array, var_name):
        """
        Specialized transformation for BIOGEOCHEMICAL variables.
        Applies tailored non-linear functions per variable type to
        maximize LE1 purity and minimize cross-contamination with LE2.
        """
        if data_array is None or len(data_array) == 0:
            return None

        scaler = RobustScaler()
        data_scaled = scaler.fit_transform(data_array.reshape(-1, 1)).flatten()

        if data_scaled.max() > data_scaled.min():
            data_scaled = (data_scaled - data_scaled.min()) / (data_scaled.max() - data_scaled.min())

        var_lower = var_name.lower()
        var_hash = abs(hash(var_name)) % 11

        if 'ph' in var_lower:
            k = 12 + var_hash * 2
            transformed = 1 / (1 + np.exp(-k * (data_scaled - 0.47)))

        elif any(nutrient in var_lower for nutrient in ['phosphate', 'nitrate', 'iron', 'silicate']):
            n = 2.2 + var_hash * 0.3
            Km = 0.05 + (var_hash * 0.02)
            transformed = data_scaled**n / (Km + data_scaled**n)

        elif any(physical in var_lower for physical in ['dissolved_oxygen', 'salinity', 'temperature']):
            k1, k2 = 6 + var_hash, 4 + var_hash * 0.5
            transformed = (0.7 / (1 + np.exp(-k1 * (data_scaled - 0.3))) +
                           0.3 / (1 + np.exp(-k2 * (data_scaled - 0.7))))

        elif any(bio in var_lower for bio in ['chlorophyll', 'phytoplankton', 'primary_productivity']):
            alpha = 4 + var_hash * 0.6
            beta = 1.2 + var_hash * 0.1
            transformed = (1 - np.exp(-alpha * data_scaled**beta))

        elif any(light in var_lower for light in ['light_bottom', 'd_a_c']):
            gamma = 3 + var_hash * 0.4
            transformed = np.exp(-gamma * (1 - data_scaled))

        else:
            k = 8 + var_hash
            transformed = (0.6 * (1 / (1 + np.exp(-k * (data_scaled - 0.4)))) +
                           0.4 * data_scaled**2)

        var_seed = GLOBAL_SEED + abs(hash(var_name + "_BIOGEOCHEM_UNIQUE")) % 1500
        np.random.seed(var_seed)
        noise = np.random.normal(0, 0.0008, len(transformed))

        transformed = np.clip(transformed + noise, 0, 1)
        return transformed.astype(np.float32)

    def anti_contamination_hydrogeo_transform(self, data_array, var_name):
        """
        Specialized transformation for HYDROGEOMORPHOLOGICAL variables.
        Applies tailored non-linear functions per variable type to
        maximize LE2 purity and minimize cross-contamination with LE1.
        """
        if data_array is None or len(data_array) == 0:
            return None

        scaler = RobustScaler()
        data_scaled = scaler.fit_transform(data_array.reshape(-1, 1)).flatten()

        if data_scaled.max() > data_scaled.min():
            data_scaled = (data_scaled - data_scaled.min()) / (data_scaled.max() - data_scaled.min())

        var_lower = var_name.lower()
        var_hash = abs(hash(var_name)) % 13

        if 'r_i_m_b' in var_lower:
            k1 = 4 + var_hash * 0.4
            k2 = 2 + var_hash * 0.3
            bio_component = 1 / (1 + np.exp(-k1 * (data_scaled - 0.4)))
            geo_component = (2/np.pi) * np.arctan(k2 * data_scaled)
            transformed = 0.6 * bio_component + 0.4 * geo_component

        elif any(bat in var_lower for bat in ['l_t_m', 'superficie']):
            power = 0.3 + var_hash * 0.1
            transformed = (1 - data_scaled)**power

        elif any(flow in var_lower for flow in ['currents_velocity', 'wind_v', 'runoff']):
            freq = 0.8 + var_hash * 0.15
            phase = var_hash * 0.5
            transformed = 0.5 * (1 + np.cos(np.pi * freq * data_scaled + phase))

        elif any(topo in var_lower for topo in ['slope', 'aspect', 'land_distance']):
            k = 2 + var_hash * 0.3
            transformed = (2/np.pi) * np.arctan(k * data_scaled)

        elif any(morph in var_lower for morph in ['r_d_m_v', 'stde']):
            transformed = np.log(1 + 9 * data_scaled) / np.log(10)

        elif any(spatial in var_lower for spatial in ['eastness', 'northness', 'river_influence']):
            k = 3 + var_hash * 0.4
            transformed = np.tanh(k * (data_scaled - 0.5)) * 0.5 + 0.5

        else:
            k1, k2 = 0.15 + var_hash * 0.05, 1.5 + var_hash * 0.2
            transformed = k1 + (1 - k1) * (data_scaled**k2) / (0.3 + data_scaled**k2)

        var_seed = GLOBAL_SEED + abs(hash(var_name + "_HYDROGEO_UNIQUE")) % 2000 + 777
        np.random.seed(var_seed)
        geo_noise = np.random.normal(0, 0.002, len(transformed))

        transformed = np.clip(transformed + geo_noise, 0, 1)
        return transformed.astype(np.float32)

    def load_category_anti_contamination(self, variables, category_idx):
        """
        Load and process a full category of variables with specialized
        anti-contamination transformations.

        category_idx = 0 → BIOGEOCHEMICAL (LE1)
        category_idx = 1 → HYDROGEOMORPHOLOGICAL (LE2)
        """
        category_names = ["BIOGEOCHEMICAL (LE1)", "HYDROGEOMORPHOLOGICAL (LE2)"]
        logger.info(f"Loading {category_names[category_idx]} with anti-contamination transforms...")

        data_dict = {}
        for var_name in variables:
            if self.df is not None and var_name in self.df.columns:
                data_series = self.df[var_name].fillna(method='bfill').fillna(method='ffill')

                if category_idx == 0:
                    transformed = self.anti_contamination_biogeochem_transform(data_series.values, var_name)
                else:
                    transformed = self.anti_contamination_hydrogeo_transform(data_series.values, var_name)

                if transformed is not None:
                    data_dict[var_name] = transformed
                    logger.info(f"   {var_name}: mean={transformed.mean():.4f}, std={transformed.std():.4f}")

        logger.info(f"{category_names[category_idx]}: {len(data_dict)} variables processed")
        return data_dict

    def extract_original_variables(self):
        """Extract untransformed original variables for MIC analysis."""
        if self.df is None:
            return None

        all_variables = biogeochemical_vars + hydrogeo_vars
        available_vars = [var for var in all_variables if var in self.df.columns]

        original_vars_df = self.df[available_vars].copy()
        for col in original_vars_df.columns:
            original_vars_df[col] = original_vars_df[col].fillna(method='bfill').fillna(method='ffill')

        return original_vars_df

# =====================================================================
# SECTION F: SPECIALIZED ENCODER AND DECODER ARCHITECTURE
# =====================================================================

def create_anti_contamination_encoder(input_layer, name, category_idx, hp):
    """
    Create a specialized encoder per data category.

    category_idx = 0 → LE1 encoder (BIOGEOCHEMICAL variables)
    category_idx = 1 → LE2 encoder (HYDROGEOMORPHOLOGICAL variables)
    """

    if category_idx == 0:  # LE1 — BIOGEOCHEMICAL ENCODER
        units1 = hp.Int(f'{name}_biogeo_units1', min_value=192, max_value=288, step=32)
        units2 = hp.Int(f'{name}_biogeo_units2', min_value=144, max_value=192, step=16)
        units3 = hp.Int(f'{name}_biogeo_units3', min_value=96,  max_value=144, step=16)
        units4 = hp.Int(f'{name}_biogeo_units4', min_value=64,  max_value=96,  step=8)
        units5 = hp.Int(f'{name}_biogeo_units5', min_value=32,  max_value=64,  step=8)

        drop1 = hp.Float(f'{name}_biogeo_drop1', min_value=0.03, max_value=0.08, step=0.01)
        drop2 = hp.Float(f'{name}_biogeo_drop2', min_value=0.05, max_value=0.12, step=0.02)
        drop3 = hp.Float(f'{name}_biogeo_drop3', min_value=0.08, max_value=0.15, step=0.02)
        drop4 = hp.Float(f'{name}_biogeo_drop4', min_value=0.1,  max_value=0.18, step=0.02)

        x = Dense(units1, activation='relu',  name=f'{name}_biogeo_layer1')(input_layer)
        x = BatchNormalization()(x)
        x = Dropout(drop1, seed=GLOBAL_SEED)(x)

        x = Dense(units2, activation='swish', name=f'{name}_biogeo_layer2')(x)
        x = BatchNormalization()(x)
        x = Dropout(drop2, seed=GLOBAL_SEED)(x)

        x = Dense(units3, activation='elu',   name=f'{name}_biogeo_layer3')(x)
        x = BatchNormalization()(x)
        x = Dropout(drop3, seed=GLOBAL_SEED)(x)

        x = Dense(units4, activation='gelu',  name=f'{name}_biogeo_layer4')(x)
        x = BatchNormalization()(x)
        x = Dropout(drop4, seed=GLOBAL_SEED)(x)

        x = Dense(units5, activation='selu',  name=f'{name}_biogeo_layer5')(x)
        x = BatchNormalization()(x)

        kernel_init = tf.keras.initializers.RandomNormal(mean=0.4, stddev=0.1, seed=GLOBAL_SEED)

    else:  # LE2 — HYDROGEOMORPHOLOGICAL ENCODER
        units1 = hp.Int(f'{name}_hydrogeo_units1', min_value=128, max_value=224, step=16)
        units2 = hp.Int(f'{name}_hydrogeo_units2', min_value=96,  max_value=160, step=8)
        units3 = hp.Int(f'{name}_hydrogeo_units3', min_value=64,  max_value=128, step=8)
        units4 = hp.Int(f'{name}_hydrogeo_units4', min_value=32,  max_value=80,  step=8)

        drop1 = hp.Float(f'{name}_hydrogeo_drop1', min_value=0.10, max_value=0.20, step=0.02)
        drop2 = hp.Float(f'{name}_hydrogeo_drop2', min_value=0.15, max_value=0.25, step=0.02)
        drop3 = hp.Float(f'{name}_hydrogeo_drop3', min_value=0.20, max_value=0.32, step=0.02)

        x = Dense(units1, activation='selu',  name=f'{name}_hydrogeo_layer1')(input_layer)
        x = BatchNormalization()(x)
        x = Dropout(drop1, seed=GLOBAL_SEED)(x)

        x = Dense(units2, activation='elu',   name=f'{name}_hydrogeo_layer2')(x)
        x = BatchNormalization()(x)
        x = Dropout(drop2, seed=GLOBAL_SEED)(x)

        x = Dense(units3, activation='relu',  name=f'{name}_hydrogeo_layer3')(x)
        x = BatchNormalization()(x)
        x = Dropout(drop3, seed=GLOBAL_SEED)(x)

        x = Dense(units4, activation='swish', name=f'{name}_hydrogeo_layer4')(x)
        x = BatchNormalization()(x)

        kernel_init = tf.keras.initializers.RandomNormal(mean=-0.6, stddev=0.12, seed=GLOBAL_SEED)

    l1_reg = hp.Float(f'{name}_l1', min_value=0.03, max_value=0.1,  step=0.01)
    l2_reg = hp.Float(f'{name}_l2', min_value=0.1,  max_value=0.25, step=0.03)
    final_dropout = hp.Float(f'{name}_final_drop', min_value=0.3, max_value=0.5, step=0.05)

    x = Dropout(final_dropout, seed=GLOBAL_SEED)(x)

    latent = Dense(1, activation='tanh',
                   name=f'LE{category_idx+1}_{name}_OPTIMIZED',
                   kernel_initializer=kernel_init,
                   kernel_regularizer=l1_l2(l1=l1_reg, l2=l2_reg))(x)

    return latent


def create_anti_contamination_decoder(latent_input, output_dim, name, category_idx, hp):
    """
    Create a specialized decoder to reconstruct data for each category.

    category_idx = 0 → LE1 decoder (BIOGEOCHEMICAL)
    category_idx = 1 → LE2 decoder (HYDROGEOMORPHOLOGICAL)
    """

    if category_idx == 0:  # BIOGEOCHEMICAL DECODER
        units1 = hp.Int(f'{name}_dec_biogeo1', min_value=32,  max_value=64,  step=8)
        units2 = hp.Int(f'{name}_dec_biogeo2', min_value=64,  max_value=96,  step=8)
        units3 = hp.Int(f'{name}_dec_biogeo3', min_value=96,  max_value=144, step=16)
        units4 = hp.Int(f'{name}_dec_biogeo4', min_value=144, max_value=192, step=16)
        units5 = hp.Int(f'{name}_dec_biogeo5', min_value=192, max_value=288, step=32)

        dropout = hp.Float(f'{name}_dec_biogeo_drop', min_value=0.03, max_value=0.1, step=0.02)

        x = Dense(units1, activation='selu',  name=f'{name}_dec1')(latent_input)
        x = BatchNormalization()(x)
        x = Dropout(dropout, seed=GLOBAL_SEED)(x)

        x = Dense(units2, activation='gelu',  name=f'{name}_dec2')(x)
        x = BatchNormalization()(x)
        x = Dropout(dropout * 0.8, seed=GLOBAL_SEED)(x)

        x = Dense(units3, activation='elu',   name=f'{name}_dec3')(x)
        x = BatchNormalization()(x)
        x = Dropout(dropout * 0.6, seed=GLOBAL_SEED)(x)

        x = Dense(units4, activation='swish', name=f'{name}_dec4')(x)
        x = BatchNormalization()(x)
        x = Dropout(dropout * 0.4, seed=GLOBAL_SEED)(x)

        x = Dense(units5, activation='relu',  name=f'{name}_dec5')(x)
        x = BatchNormalization()(x)

    else:  # HYDROGEOMORPHOLOGICAL DECODER
        units1 = hp.Int(f'{name}_dec_hydrogeo1', min_value=32,  max_value=80,  step=8)
        units2 = hp.Int(f'{name}_dec_hydrogeo2', min_value=64,  max_value=128, step=8)
        units3 = hp.Int(f'{name}_dec_hydrogeo3', min_value=96,  max_value=160, step=8)
        units4 = hp.Int(f'{name}_dec_hydrogeo4', min_value=128, max_value=224, step=16)

        dropout = hp.Float(f'{name}_dec_hydrogeo_drop', min_value=0.10, max_value=0.22, step=0.03)

        x = Dense(units1, activation='swish', name=f'{name}_dec1')(latent_input)
        x = BatchNormalization()(x)
        x = Dropout(dropout, seed=GLOBAL_SEED)(x)

        x = Dense(units2, activation='relu',  name=f'{name}_dec2')(x)
        x = BatchNormalization()(x)
        x = Dropout(dropout * 0.9, seed=GLOBAL_SEED)(x)

        x = Dense(units3, activation='elu',   name=f'{name}_dec3')(x)
        x = BatchNormalization()(x)
        x = Dropout(dropout * 0.8, seed=GLOBAL_SEED)(x)

        x = Dense(units4, activation='selu',  name=f'{name}_dec4')(x)
        x = BatchNormalization()(x)

    output = Dense(output_dim, activation='sigmoid', name=f'output_{name}')(x)
    return output

# =====================================================================
# SECTION G: ULTRA-OPTIMIZED DUAL-OBJECTIVE REGULARIZATION
# =====================================================================

def create_optimized_dual_regularizer(hp, biogeochem_data_tensor, hydrogeo_data_tensor):
    """
    Create a regularizer implementing the balanced dual-objective strategy:
      - LE1 specialization on BIOGEOCHEMICAL variables
      - LE2 specialization on HYDROGEOMORPHOLOGICAL variables (priority)
      - Soft orthogonality constraint between LE1 and LE2
    """

    ortho_weight                 = hp.Float('orthogonal_weight',              min_value=500,  max_value=1500, step=150)
    purity_weight                = hp.Float('purity_weight',                  min_value=3000, max_value=5500, step=300)
    cross_contamination_penalty  = hp.Float('cross_contamination_penalty',    min_value=2200, max_value=4000, step=300)
    le2_boost_weight             = hp.Float('le2_boost_weight',               min_value=4000, max_value=6500, step=300)
    variance_balance_weight      = hp.Float('variance_balance_weight',        min_value=400,  max_value=1200, step=150)
    orthogonal_separation_weight = hp.Float('orthogonal_separation_weight',   min_value=600,  max_value=1800, step=200)
    specialization_reward_weight = hp.Float('specialization_reward_weight',   min_value=2000, max_value=3800, step=250)
    le2_position_boost           = hp.Float('le2_position_boost',             min_value=2.5,  max_value=4.5,  step=0.3)
    le2_hydrogeo_boost           = hp.Float('le2_hydrogeo_boost',             min_value=1500, max_value=3000, step=200)

    def optimized_dual_loss(latent_concat):
        le1, le2 = latent_concat[:, 0:1], latent_concat[:, 1:2]

        # PRIORITY 1: DIRECT SPECIALIZATION REWARD
        biogeo_mean   = tf.reduce_mean(biogeochem_data_tensor, axis=1, keepdims=True)
        hydrogeo_mean = tf.reduce_mean(hydrogeo_data_tensor,   axis=1, keepdims=True)

        le1_biogeo_alignment   = tf.abs(tf.reduce_mean(tf.multiply(le1, biogeo_mean)))
        le2_hydrogeo_alignment = tf.abs(tf.reduce_mean(tf.multiply(le2, hydrogeo_mean)))

        specialization_reward = -(le1_biogeo_alignment + le2_hydrogeo_alignment * 1.5) * specialization_reward_weight

        # PRIORITY 2: LE2 ULTRA-MAXIMUM BOOST (hydrogeomorphological recovery)
        le1_mean, le2_mean = tf.reduce_mean(le1), tf.reduce_mean(le2)
        var_le2 = tf.math.reduce_variance(le2)

        le2_variance_boost       = -var_le2                       * le2_boost_weight * 2.8
        le2_mean_boost           = -tf.abs(le2_mean + 0.15)       * le2_boost_weight * 2.2
        le2_activity_boost       = -tf.reduce_mean(tf.abs(le2))   * le2_boost_weight * 1.5
        le2_spread_boost         = -tf.math.reduce_std(le2)       * le2_boost_weight * 1.2
        le2_hydrogeo_specialization = -tf.abs(le2_mean + 0.3)     * le2_hydrogeo_boost

        # PRIORITY 3: POSITION SEPARATION (LE1 positive zone, LE2 negative zone)
        le1_position_penalty = tf.maximum(0.0, 0.08 - le1_mean) * purity_weight * 1.2
        le2_position_penalty = tf.maximum(0.0, le2_mean + 0.08) * purity_weight * le2_position_boost
        position_separation  = le1_position_penalty + le2_position_penalty

        # PRIORITY 4: ANTI-CROSS-CONTAMINATION
        le1_hydrogeo_corr = tf.abs(tf.reduce_mean(tf.multiply(le1, hydrogeo_mean)))
        le2_biogeo_corr   = tf.abs(tf.reduce_mean(tf.multiply(le2, biogeo_mean)))
        cross_contamination = (le1_hydrogeo_corr + le2_biogeo_corr * 3.0) * cross_contamination_penalty

        # PRIORITY 5: SOFT ORTHOGONALITY (reduced 80% vs purity)
        corr_12             = tf.reduce_mean(le1 * le2)
        orthogonality_loss  = tf.square(corr_12) * ortho_weight * 0.2
        dot_product_penalty = tf.square(tf.reduce_sum(le1 * le2)) * orthogonal_separation_weight * 0.3

        # PRIORITY 6: MINIMUM VARIANCE BALANCE
        var_le1 = tf.math.reduce_variance(le1)
        variance_ratio           = var_le1 / (var_le2 + 1e-8)
        variance_balance_penalty = tf.square(variance_ratio - 1.0) * variance_balance_weight * 0.4

        # ANTI-COLLAPSE: penalize variance collapse
        variance_penalty = -(var_le1 + var_le2) * 800

        total_optimized_loss = (
            le2_variance_boost          +
            le2_mean_boost              +
            le2_activity_boost          +
            le2_spread_boost            +
            le2_hydrogeo_specialization +
            specialization_reward       +
            position_separation         +
            cross_contamination         +
            orthogonality_loss          +
            dot_product_penalty         +
            variance_balance_penalty    +
            variance_penalty
        )

        return total_optimized_loss

    return optimized_dual_loss

# =====================================================================
# SECTION H: FULL DUAL OPTIMIZED MODEL CONSTRUCTION
# =====================================================================

def build_optimized_dual_autoencoder(hp):
    """
    Build the full model integrating all specialized components.

    Inputs:
      - input_biogeochem  → feeds LE1 encoder (BIOGEOCHEMICAL)
      - input_hydrogeo    → feeds LE2 encoder (HYDROGEOMORPHOLOGICAL)
    Outputs:
      - Reconstructed biogeochemical variables
      - Reconstructed hydrogeomorphological variables
    """

    global global_data_shapes

    input_biogeochem = Input(shape=(global_data_shapes['biogeochem'],), name='input_biogeochem')
    input_hydrogeo   = Input(shape=(global_data_shapes['hydrogeo'],),   name='input_hydrogeo')

    # Encoders
    latent_biogeochem = create_anti_contamination_encoder(input_biogeochem, 'biogeochem', 0, hp)
    latent_hydrogeo   = create_anti_contamination_encoder(input_hydrogeo,   'hydrogeo',   1, hp)

    # Concatenate latent space [LE1, LE2]
    latent_vector = concatenate([latent_biogeochem, latent_hydrogeo])

    # Apply dual-objective regularizer
    optimized_dual_regularizer = create_optimized_dual_regularizer(hp, input_biogeochem, input_hydrogeo)
    specialized_latent = Lambda(lambda x: x, activity_regularizer=optimized_dual_regularizer)(latent_vector)

    # Decoders
    output_biogeochem = create_anti_contamination_decoder(latent_biogeochem, global_data_shapes['biogeochem'], 'biogeochem', 0, hp)
    output_hydrogeo   = create_anti_contamination_decoder(latent_hydrogeo,   global_data_shapes['hydrogeo'],   'hydrogeo',   1, hp)

    model = Model(
        inputs=[input_biogeochem, input_hydrogeo],
        outputs=[output_biogeochem, output_hydrogeo]
    )

    learning_rate = hp.Float('learning_rate', min_value=5e-5, max_value=8e-4, sampling='LOG')
    optimizer = Adam(learning_rate=learning_rate, clipnorm=0.2)

    biogeo_weight   = hp.Float('biogeo_weight',   min_value=1.2,  max_value=2.2,  step=0.2)
    hydrogeo_weight = hp.Float('hydrogeo_weight', min_value=14.0, max_value=22.0, step=1.5)

    model.compile(
        optimizer=optimizer,
        loss={'output_biogeochem': 'mse', 'output_hydrogeo': 'mse'},
        loss_weights={'output_biogeochem': biogeo_weight, 'output_hydrogeo': hydrogeo_weight}
    )

    return model

# =====================================================================
# SECTION I: SPECIALIZED MONITORING CALLBACKS
# =====================================================================

class OptimizedDualCallback(Callback):
    """
    Specialized callback to monitor dual-objective progress in real time.
    Tracks LE2 (hydrogeomorphological) purity recovery alongside
    LE1 (biogeochemical) stability and overall orthogonality.
    """

    def __init__(self, validation_data, biogeochem_data, hydrogeo_data, check_interval=5):
        super().__init__()
        self.validation_data        = validation_data
        self.biogeochem_data        = biogeochem_data
        self.hydrogeo_data          = hydrogeo_data
        self.check_interval         = check_interval
        self.best_le2_purity        = 0.0
        self.le2_improvement_epochs = []
        self.target_le2_reached     = False

    def on_epoch_end(self, epoch, logs=None):
        if epoch % self.check_interval == 0 and epoch > 0:
            try:
                hydrogeo_loss = logs.get('val_output_hydrogeo_loss', 0)
                biogeo_loss   = logs.get('val_output_biogeochem_loss', 0)

                inputs = self.model.inputs
                le_outputs = [
                    self.model.get_layer('LE1_biogeochem_OPTIMIZED').output,
                    self.model.get_layer('LE2_hydrogeo_OPTIMIZED').output
                ]
                temp_extractor = Model(inputs, concatenate(le_outputs))
                latent_pred = temp_extractor.predict(
                    [self.validation_data[0][:100], self.validation_data[1][:100]], verbose=0
                )

                le1_sample, le2_sample = latent_pred[:, 0], latent_pred[:, 1]

                le2_mean     = np.mean(le2_sample)
                le2_std      = np.std(le2_sample)
                le2_activity = np.mean(np.abs(le2_sample))
                le2_range    = np.max(le2_sample) - np.min(le2_sample)

                estimated_le2_purity = (le2_activity * le2_std * le2_range) ** 0.5

                if estimated_le2_purity > self.best_le2_purity:
                    self.best_le2_purity = estimated_le2_purity
                    self.le2_improvement_epochs.append(epoch)

                if estimated_le2_purity > 0.5 and not self.target_le2_reached:
                    self.target_le2_reached = True
                    logger.info(f"LE2 TARGET REACHED at epoch {epoch}!")

                corr_sample   = np.corrcoef(le1_sample, le2_sample)[0, 1]
                balance_ratio = hydrogeo_loss / (biogeo_loss + 1e-6)

                logger.info(f"Epoch {epoch} — DUAL OPTIMIZED:")
                logger.info(f"  LE2 (Hydrogeo): mean={le2_mean:.3f}, std={le2_std:.3f}, activity={le2_activity:.3f}")
                logger.info(f"  LE2 (Hydrogeo): range={le2_range:.3f}, purity_proxy={estimated_le2_purity:.4f}")
                logger.info(f"  BALANCE: Ratio={balance_ratio:.2f}, Hydrogeo_loss={hydrogeo_loss:.4f}")
                logger.info(f"  ORTHOGONALITY: r={corr_sample:.3f}")
                logger.info(f"  LE2 RECORD: Best={self.best_le2_purity:.4f}")

            except Exception as e:
                logger.warning(f"Monitoring error at epoch {epoch}: {e}")


class MemoryManagementCallback(Callback):
    """Callback for memory management during long training runs."""
    def on_epoch_end(self, epoch, logs=None):
        if epoch % 3 == 0:
            gc.collect()


def plot_optimized_training(history, output_dir, timestamp):
    """Generate specialized charts for dual-objective training analysis."""
    try:
        plt.figure(figsize=(24, 16))

        plt.subplot(3, 4, 1)
        plt.plot(history.history['loss'],     label='Train', linewidth=2, color='blue')
        plt.plot(history.history['val_loss'], label='Val',   linewidth=2, color='red')
        plt.title('Total Loss')
        plt.legend()
        plt.grid(True, alpha=0.3)

        plt.subplot(3, 4, 2)
        if 'output_hydrogeo_loss' in history.history:
            plt.plot(history.history['val_output_hydrogeo_loss'],   '-', color='green',  linewidth=4, label='Hydrogeo Val (LE2)')
        if 'output_biogeochem_loss' in history.history:
            plt.plot(history.history['val_output_biogeochem_loss'], '-', color='purple', linewidth=2, label='Biogeochem Val (LE1)')
        plt.title('Purity Balance — LE2 should dominate')
        plt.legend()
        plt.grid(True, alpha=0.3)

        plt.subplot(3, 4, 3)
        if 'val_output_hydrogeo_loss' in history.history and 'val_output_biogeochem_loss' in history.history:
            hydrogeo_losses = np.array(history.history['val_output_hydrogeo_loss'])
            biogeo_losses   = np.array(history.history['val_output_biogeochem_loss'])
            ratios = hydrogeo_losses / (biogeo_losses + 1e-8)
            plt.plot(ratios, color='red', linewidth=3)
            plt.axhline(y=15, color='green',  linestyle='--', alpha=0.8, label='Target >15 for LE2')
            plt.axhline(y=10, color='orange', linestyle='--', alpha=0.6, label='Minimum 10')
            plt.title('Hydrogeo/Biogeochem Ratio — Higher = Better LE2')
            plt.legend()
            plt.grid(True, alpha=0.3)

        plt.subplot(3, 4, 4)
        if 'output_hydrogeo_loss' in history.history:
            plt.plot(history.history['output_hydrogeo_loss'],   '-', color='darkgreen', linewidth=2, label='Hydrogeo Train')
        if 'output_biogeochem_loss' in history.history:
            plt.plot(history.history['output_biogeochem_loss'], '-', color='darkred',   linewidth=2, label='Biogeochem Train')
        plt.title('Training Losses')
        plt.legend()
        plt.grid(True, alpha=0.3)

        plt.subplot(3, 4, 5)
        epochs = range(1, len(history.history['loss']) + 1)
        plt.plot(epochs, history.history['loss'], 'b-', label='Total Loss', linewidth=2)
        if 'val_output_hydrogeo_loss' in history.history:
            plt.plot(epochs, history.history['val_output_hydrogeo_loss'], 'g-', linewidth=3, label='Hydrogeo Val')
        plt.title('Comparative Evolution')
        plt.legend()
        plt.grid(True, alpha=0.3)

        plt.tight_layout()
        plot_path = os.path.join(output_dir, f'ECOSAIC_training_{timestamp}.png')
        plt.savefig(plot_path, dpi=300, bbox_inches='tight')
        logger.info(f"Charts saved: {plot_path}")
        plt.show()
        return plot_path

    except Exception as e:
        logger.error(f"Error generating charts: {e}")
        return None

# =====================================================================
# SECTION J: MAIN ORCHESTRATION FUNCTION
# =====================================================================

def main_optimized_dual_autoencoder():
    """
    Main function executing the complete dual-objective training pipeline.

    LE1 → BIOGEOCHEMICAL latent embedding
    LE2 → HYDROGEOMORPHOLOGICAL latent embedding (recovery priority)
    """

    logger.info("ECOSAIC — OPTIMIZED DUAL AUTOENCODER")
    logger.info("="*80)

    mic_available  = setup_signed_realistic_mic()
    output_dir     = "data/outputs"
    checkpoint_dir = "data/checkpoints"
    os.makedirs(output_dir,     exist_ok=True)
    os.makedirs(checkpoint_dir, exist_ok=True)

    try:
        # PHASE 1: DATA LOADING AND PROCESSING
        processor = OptimizedDualProcessor(excel_file_path)
        if not processor.load_excel_data():
            return None

        biogeo_data   = processor.load_category_anti_contamination(biogeochemical_vars, 0)  # LE1
        hydrogeo_data = processor.load_category_anti_contamination(hydrogeo_vars,       1)  # LE2

        biogeo_arrays   = [biogeo_data[var]  for var in biogeochemical_vars if var in biogeo_data]
        hydrogeo_arrays = [hydrogeo_data[var] for var in hydrogeo_vars       if var in hydrogeo_data]

        if not biogeo_arrays or not hydrogeo_arrays:
            logger.error("Missing data categories")
            return None

        biogeo_matrix   = np.column_stack(biogeo_arrays).astype(np.float32)
        hydrogeo_matrix = np.column_stack(hydrogeo_arrays).astype(np.float32)

        logger.info(f"Biogeochemical (LE1)        : {biogeo_matrix.shape}")
        logger.info(f"Hydrogeomorphological (LE2) : {hydrogeo_matrix.shape}")

        global global_data_shapes
        global_data_shapes = {
            'biogeochem': biogeo_matrix.shape[1],
            'hydrogeo':   hydrogeo_matrix.shape[1]
        }

        n_samples = biogeo_matrix.shape[0]
        train_idx, val_idx = train_test_split(np.arange(n_samples), test_size=0.2, random_state=GLOBAL_SEED)

        train_data = {
            'x': [biogeo_matrix[train_idx], hydrogeo_matrix[train_idx]],
            'y': {
                'output_biogeochem': biogeo_matrix[train_idx],
                'output_hydrogeo':   hydrogeo_matrix[train_idx]
            }
        }

        val_data = {
            'x': [biogeo_matrix[val_idx], hydrogeo_matrix[val_idx]],
            'y': {
                'output_biogeochem': biogeo_matrix[val_idx],
                'output_hydrogeo':   hydrogeo_matrix[val_idx]
            }
        }

        # PHASE 2: HYPERPARAMETER TUNING
        logger.info("Hyperparameter search (LE2 focus)...")

        tuner = kt.RandomSearch(
            build_optimized_dual_autoencoder,
            objective='val_loss',
            max_trials=30,
            executions_per_trial=1,
            directory=os.path.join(checkpoint_dir, 'tuner_optimized_dual'),
            project_name='optimized_dual',
            seed=GLOBAL_SEED
        )

        tuner_callbacks = [
            EarlyStopping(monitor='val_loss', patience=20, restore_best_weights=True),
            MemoryManagementCallback()
        ]

        tuner.search(
            x=train_data['x'],
            y=train_data['y'],
            epochs=45,
            batch_size=32,
            validation_data=(val_data['x'], val_data['y']),
            callbacks=tuner_callbacks,
            verbose=1
        )

        best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
        logger.info("Best hyperparameters found:")

        critical_params = [
            'le2_boost_weight', 'hydrogeo_weight', 'le2_position_boost',
            'le2_hydrogeo_boost', 'specialization_reward_weight', 'orthogonal_weight'
        ]
        for param in critical_params:
            if param in best_hps.values:
                logger.info(f"   {param}: {best_hps.values[param]}")

        # PHASE 3: FINAL MODEL TRAINING
        best_model = build_optimized_dual_autoencoder(best_hps)
        logger.info("Training final model...")

        callbacks = [
            EarlyStopping(monitor='val_loss', patience=55, restore_best_weights=True, verbose=1),
            ReduceLROnPlateau(monitor='loss', factor=0.2, patience=30, min_lr=1e-8, verbose=1),
            MemoryManagementCallback(),
            OptimizedDualCallback(val_data['x'], biogeo_data, hydrogeo_data, check_interval=4),
            ModelCheckpoint(
                filepath=os.path.join(checkpoint_dir, 'best_model.keras'),
                save_best_only=True,
                monitor='val_loss'
            )
        ]

        history = best_model.fit(
            x=train_data['x'],
            y=train_data['y'],
            validation_data=(val_data['x'], val_data['y']),
            epochs=220,
            batch_size=32,
            callbacks=callbacks,
            verbose=1
        )

        logger.info("Training complete")

        # PHASE 4: RESULT EXTRACTION
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        plot_path = plot_optimized_training(history, output_dir, timestamp)

        def create_optimized_extractor(trained_model):
            inputs = trained_model.inputs
            le_outputs = [
                trained_model.get_layer('LE1_biogeochem_OPTIMIZED').output,
                trained_model.get_layer('LE2_hydrogeo_OPTIMIZED').output
            ]
            latent_concat = concatenate(le_outputs)
            return Model(inputs, latent_concat)

        encoder         = create_optimized_extractor(best_model)
        latent_features = encoder.predict([biogeo_matrix, hydrogeo_matrix], batch_size=64, verbose=1)

        latent_df = pd.DataFrame(
            latent_features,
            columns=["LE1_Biogeochemical", "LE2_Hydrogeomorphological"]
        )

        le1_values, le2_values = latent_features[:, 0], latent_features[:, 1]
        le1_mean, le2_mean     = np.mean(le1_values), np.mean(le2_values)
        separation_achieved    = le1_mean - le2_mean

        original_vars_df = processor.extract_original_variables()
        if original_vars_df is None:
            logger.error("Could not extract original variables")
            return None

        final_df = pd.concat([
            processor.coordinates.reset_index(drop=True),
            latent_df.reset_index(drop=True),
            original_vars_df.reset_index(drop=True)
        ], axis=1)

        # PHASE 5: MIC PURITY METRICS
        original_data_combined = {}
        original_data_combined.update({f"BIOGEOCHEM_{k}": v for k, v in biogeo_data.items()})
        original_data_combined.update({f"HYDROGEO_{k}":   v for k, v in hydrogeo_data.items()})
        original_df_mic = pd.DataFrame(original_data_combined)

        mic_matrix = calculate_signed_realistic_mic_matrix(latent_df, original_df_mic)

        biogeo_cols   = [col for col in mic_matrix.columns if col.startswith('BIOGEOCHEM_')]
        hydrogeo_cols = [col for col in mic_matrix.columns if col.startswith('HYDROGEO_')]

        # LE1 PURITY
        le1_biogeo_mic   = mic_matrix.loc['LE1_Biogeochemical',       biogeo_cols].abs().mean()
        le1_hydrogeo_mic = mic_matrix.loc['LE1_Biogeochemical',       hydrogeo_cols].abs().mean()
        le1_specialization_ratio = le1_biogeo_mic / (le1_hydrogeo_mic + 1e-6)
        le1_purity = max(0, 1 - (le1_hydrogeo_mic / (le1_biogeo_mic + 1e-6)))

        # LE2 PURITY
        le2_hydrogeo_mic = mic_matrix.loc['LE2_Hydrogeomorphological', hydrogeo_cols].abs().mean()
        le2_biogeo_mic   = mic_matrix.loc['LE2_Hydrogeomorphological', biogeo_cols].abs().mean()
        le2_specialization_ratio = le2_hydrogeo_mic / (le2_biogeo_mic + 1e-6)
        le2_purity = max(0, 1 - (le2_biogeo_mic / (le2_hydrogeo_mic + 1e-6)))

        # PHASE 6: ORTHOGONALITY
        corr_le1_le2        = np.corrcoef(le1_values, le2_values)[0, 1]
        orthogonality_score = max(0, 1 - abs(corr_le1_le2))
        var_le1, var_le2    = np.var(le1_values), np.var(le2_values)
        variance_balance    = min(var_le1, var_le2) / max(var_le1, var_le2)

        # PHASE 7: COMBINED SCORE
        optimized_success = (
            (le1_purity          * 0.20) +
            (le2_purity          * 0.50) +
            (orthogonality_score * 0.15) +
            (variance_balance    * 0.15)
        )

        # PHASE 8: STATUS
        le2_recovery_level = ("EXCELLENT"    if le2_purity > 0.35 else
                              "GOOD"         if le2_purity > 0.25 else
                              "PROGRESS"     if le2_purity > 0.15 else
                              "INSUFFICIENT")

        orthogonality_status = ("EXCELLENT"  if abs(corr_le1_le2) < 0.18 else
                                "GOOD"       if abs(corr_le1_le2) < 0.25 else
                                "ACCEPTABLE" if abs(corr_le1_le2) < 0.35 else
                                "IMPROVABLE")

        purity_status = ("RECOVERED"    if (le1_purity > 0.35 and le2_purity > 0.30) else
                         "LE2 PROGRESS" if le2_purity > 0.15 else
                         "LE2 CRITICAL")

        optimized_status = ("SUCCESSFUL"         if optimized_success > 0.65 else
                            "IN PROGRESS"        if optimized_success > 0.50 else
                            "REQUIRES MORE BOOST")

        # PHASE 9: SAVE OUTPUTS
        # ── Latent embeddings + original variables ───────────────────
        latent_path = os.path.join(output_dir, f'LE_DUAL_OPTIMIZED_{timestamp}.csv')
        final_df.to_csv(latent_path, index=False)

        # ── MIC / correlation matrix ─────────────────────────────────
        mic_suffix = "MIC" if mic_available else "CORR"
        mic_path   = os.path.join(output_dir, f'matrix_{mic_suffix}_DUAL_OPTIMIZED_{timestamp}.csv')
        mic_matrix.to_csv(mic_path, index=True)

        # ── Model weights ────────────────────────────────────────────
        model_path   = os.path.join(checkpoint_dir, f'model_dual_optimized_{timestamp}.keras')
        encoder_path = os.path.join(checkpoint_dir, f'encoder_dual_optimized_{timestamp}.keras')
        best_model.save(model_path)
        encoder.save(encoder_path)

        best_val_loss = min(history.history['val_loss'])
        total_epochs  = len(history.history['loss'])
        best_epoch    = np.argmin(history.history['val_loss']) + 1

        # PHASE 10: FINAL REPORT
        logger.info("="*80)
        logger.info("ECOSAIC — FINAL RESULT")
        logger.info("="*80)
        logger.info(f"Best val_loss : {best_val_loss:.8f}")
        logger.info(f"Best epoch    : {best_epoch}/{total_epochs}")
        logger.info("")
        logger.info("LE2 — HYDROGEOMORPHOLOGICAL:")
        logger.info(f"  Purity  : {le2_purity:.4f}  (target >0.30)")
        logger.info(f"  Ratio   : {le2_specialization_ratio:.2f}")
        logger.info(f"  Level   : {le2_recovery_level}")
        logger.info("")
        logger.info("LE1 — BIOGEOCHEMICAL:")
        logger.info(f"  Purity  : {le1_purity:.4f}  (target >0.35)")
        logger.info(f"  Ratio   : {le1_specialization_ratio:.2f}")
        logger.info("")
        logger.info("ORTHOGONALITY:")
        logger.info(f"  LE1–LE2 correlation : {corr_le1_le2:.4f}  (target <0.25)")
        logger.info(f"  Score               : {orthogonality_score:.4f}")
        logger.info(f"  Status              : {orthogonality_status}")
        logger.info("")
        logger.info(f"PURITY STATUS    : {purity_status}")
        logger.info(f"OPTIMIZED SCORE  : {optimized_success:.4f}")
        logger.info(f"FINAL STATUS     : {optimized_status}")
        logger.info("="*80)
        logger.info(f"Outputs saved to : {output_dir}")
        logger.info(f"  Latent file  : {latent_path}")
        logger.info(f"  Matrix file  : {mic_path}")
        logger.info(f"  Model        : {model_path}")
        logger.info(f"  Encoder      : {encoder_path}")
        logger.info("="*80)

        return {
            'final_df':      final_df,
            'latent_df':     latent_df,
            'mic_matrix':    mic_matrix,
            'best_val_loss': best_val_loss,
            'total_epochs':  total_epochs,
            'best_epoch':    best_epoch,
            'separation_metrics': {
                'le1_mean':            le1_mean,
                'le2_mean':            le2_mean,
                'separation_achieved': separation_achieved,
                'le1_le2_corr':        corr_le1_le2,
                'orthogonality_score': orthogonality_score
            },
            'optimized_metrics': {
                'le1_purity':               le1_purity,
                'le1_specialization_ratio': le1_specialization_ratio,
                'le2_purity':               le2_purity,
                'le2_specialization_ratio': le2_specialization_ratio,
                'le2_recovery_level':       le2_recovery_level,
                'optimized_success':        optimized_success,
                'purity_status':            purity_status,
                'orthogonality_status':     orthogonality_status,
                'optimized_status':         optimized_status
            },
            'variance_metrics': {
                'le1_variance':     var_le1,
                'le2_variance':     var_le2,
                'variance_balance': variance_balance
            },
            'latent_path':  latent_path,
            'mic_path':     mic_path,
            'model_path':   model_path,
            'encoder_path': encoder_path,
            'plot_path':    plot_path,
            'model':        best_model,
            'encoder':      encoder,
            'hyperparameters': best_hps.values,
            'status': 'SUCCESS — ECOSAIC DUAL OPTIMIZED'
        }

    except Exception as e:
        logger.error(f"CRITICAL ERROR: {e}")
        import traceback
        logger.error(traceback.format_exc())
        tf.keras.backend.clear_session()
        gc.collect()
        raise

# =====================================================================
# SECTION K: MAIN EXECUTION
# =====================================================================

if __name__ == "__main__":
    print("ECOSAIC — OPTIMIZED DUAL AUTOENCODER")
    print("="*80)
    print("CATEGORY DEFINITIONS:")
    print("  LE1 → BIOGEOCHEMICAL        : physicochemical + biological variables")
    print("  LE2 → HYDROGEOMORPHOLOGICAL : hydrological + geomorphological variables")
    print("")
    print("INPUT  : data/inputs/seafloor_environmental_variables.xlsx")
    print("OUTPUTS: data/outputs/")
    print("  LE_DUAL_OPTIMIZED_<timestamp>.csv")
    print("  matrix_MIC_DUAL_OPTIMIZED_<timestamp>.csv   (or CORR if MIC unavailable)")
    print("  ECOSAIC_training_<timestamp>.png")
    print("CHECKPOINTS: data/checkpoints/")
    print("  model_dual_optimized_<timestamp>.keras")
    print("  encoder_dual_optimized_<timestamp>.keras")
    print("="*80)

    results = main_optimized_dual_autoencoder()

    if results:
        print(f"\nCOMPLETED: {results['status']}")
        print(f"Val loss  : {results['best_val_loss']:.8f}")
        print(f"Best epoch: {results['best_epoch']}/{results['total_epochs']}")

        optimized  = results['optimized_metrics']
        separation = results['separation_metrics']
        variance   = results['variance_metrics']

        print(f"\nLE2 — HYDROGEOMORPHOLOGICAL:")
        le2_symbol = "OK" if optimized['le2_purity'] > 0.30 else "PROGRESS" if optimized['le2_purity'] > 0.15 else "CRITICAL"
        print(f"  Purity : {optimized['le2_purity']:.4f}  [{le2_symbol}]")
        print(f"  Level  : {optimized['le2_recovery_level']}")

        print(f"\nLE1 — BIOGEOCHEMICAL:")
        le1_symbol = "OK" if optimized['le1_purity'] > 0.35 else "REVIEW"
        print(f"  Purity : {optimized['le1_purity']:.4f}  [{le1_symbol}]")

        print(f"\nORTHOGONALITY:")
        corr_symbol = "OK" if abs(separation['le1_le2_corr']) < 0.25 else "PROGRESS" if abs(separation['le1_le2_corr']) < 0.35 else "IMPROVABLE"
        print(f"  Correlation: {separation['le1_le2_corr']:.4f}  [{corr_symbol}]")
        print(f"  Status     : {optimized['orthogonality_status']}")

        print(f"\nOVERALL:")
        print(f"  Score : {optimized['optimized_success']:.4f}")
        print(f"  Status: {optimized['optimized_status']}")
        print(f"\nOutputs → {results['latent_path']}")
        print(f"Matrix  → {results['mic_path']}")

    else:
        print("Error during execution")
```

# 2. Specialization of latent components LE1 and LE2 evaluated by MIC

```python
import os
import re
import glob
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

# =====================================================================
# FILE DISCOVERY — reads the LATEST outputs from the autoencoder.
#
# The autoencoder (main_optimized_dual_autoencoder) writes two files:
#
#   LE_DUAL_OPTIMIZED_{timestamp}.csv          ← latent embeddings
#   matriz_CORR_DUAL_OPTIMIZED_{timestamp}.csv ← Spearman correlation
#   matriz_MIC_DUAL_OPTIMIZED_{timestamp}.csv  ← MIC (if rpy2 available)
#
# Both files live in:  data/outputs/
#
# This script auto-discovers the most recent pair so you never have to
# manually edit TIMESTAMP again.  You can still override by setting
# TIMESTAMP manually below (leave as None for auto-discovery).
# =====================================================================

OUTPUT_DIR = "data/outputs"
TIMESTAMP  = None   # ← set to e.g. "20260223_161913" to pin a specific run

# ── Auto-discover the latest run ──────────────────────────────────────
def find_latest_autoencoder_outputs(output_dir: str, timestamp: str | None):
    """
    Returns (corr_path, latent_path, resolved_timestamp).

    Search strategy
    ---------------
    1. If ``timestamp`` is given, look for both files with that exact stamp.
    2. Otherwise, enumerate all LE_DUAL_OPTIMIZED_*.csv files, extract their
       timestamps, and pick the most recent pair where *both* files exist.

    File-name variants accepted
    ---------------------------
    Latent  : LE_DUAL_OPTIMIZED_{ts}.csv
    Corr    : matriz_CORR_DUAL_OPTIMIZED_{ts}.csv
              matrix_CORR_DUAL_OPTIMIZED_{ts}.csv   (old naming)
              matriz_MIC_DUAL_OPTIMIZED_{ts}.csv
              matrix_MIC_DUAL_OPTIMIZED_{ts}.csv    (old naming)
    """

    # Helper: given a timestamp, try every known pattern for the corr/MIC file
    def _find_corr(ts: str):
        patterns = [
            f"matriz_CORR_DUAL_OPTIMIZED_{ts}.csv",
            f"matrix_CORR_DUAL_OPTIMIZED_{ts}.csv",          # old prefix
            f"matriz_MIC_DUAL_OPTIMIZED_{ts}.csv",
            f"matrix_MIC_DUAL_OPTIMIZED_{ts}.csv",           # old prefix
            # legacy names that included _R_i_m_b_
            f"matrix_CORR_DUAL_OPTIMIZED_R_i_m_b_{ts}.csv",
            f"matriz_CORR_DUAL_OPTIMIZED_R_i_m_b_{ts}.csv",
        ]
        for p in patterns:
            fp = os.path.join(output_dir, p)
            if os.path.exists(fp):
                return fp
        return None

    def _find_latent(ts: str):
        patterns = [
            f"LE_DUAL_OPTIMIZED_{ts}.csv",
            # legacy name that included _R_i_m_b_
            f"LE_DUAL_OPTIMIZED_R_i_m_b_{ts}.csv",
        ]
        for p in patterns:
            fp = os.path.join(output_dir, p)
            if os.path.exists(fp):
                return fp
        return None

    # ── Case 1: explicit timestamp ────────────────────────────────────
    if timestamp is not None:
        corr_fp   = _find_corr(timestamp)
        latent_fp = _find_latent(timestamp)
        if corr_fp is None:
            raise FileNotFoundError(
                f"Correlation/MIC file not found for timestamp '{timestamp}' "
                f"in '{output_dir}'.\n"
                f"Expected one of:\n"
                f"  matriz_CORR_DUAL_OPTIMIZED_{timestamp}.csv\n"
                f"  matriz_MIC_DUAL_OPTIMIZED_{timestamp}.csv"
            )
        if latent_fp is None:
            raise FileNotFoundError(
                f"Latent-embedding file not found for timestamp '{timestamp}' "
                f"in '{output_dir}'.\n"
                f"Expected: LE_DUAL_OPTIMIZED_{timestamp}.csv"
            )
        return corr_fp, latent_fp, timestamp

    # ── Case 2: auto-discover the most recent complete pair ───────────
    le_glob = glob.glob(os.path.join(output_dir, "LE_DUAL_OPTIMIZED_*.csv"))
    if not le_glob:
        raise FileNotFoundError(
            f"No 'LE_DUAL_OPTIMIZED_*.csv' files found in '{output_dir}'.\n"
            f"Run the autoencoder first to generate outputs."
        )

    # Extract timestamps (YYYYmmdd_HHMMSS) from file names
    ts_pattern = re.compile(r"LE_DUAL_OPTIMIZED_(?:R_i_m_b_)?(\d{8}_\d{6})\.csv$")
    candidates = []
    for fp in le_glob:
        m = ts_pattern.search(os.path.basename(fp))
        if m:
            candidates.append(m.group(1))

    if not candidates:
        raise FileNotFoundError(
            f"Found LE_DUAL_OPTIMIZED_*.csv files but could not parse timestamps from:\n"
            + "\n".join(le_glob)
        )

    # Sort descending (latest first)
    candidates.sort(reverse=True)

    for ts in candidates:
        corr_fp   = _find_corr(ts)
        latent_fp = _find_latent(ts)
        if corr_fp and latent_fp:
            return corr_fp, latent_fp, ts

    raise FileNotFoundError(
        f"Found latent files but could not match them with a correlation/MIC file.\n"
        f"Timestamps found: {candidates}\n"
        f"Check '{output_dir}' for the corresponding 'matriz_CORR_*' or 'matriz_MIC_*' files."
    )


# ── Resolve paths ─────────────────────────────────────────────────────
file_path, latent_file_path, resolved_ts = find_latest_autoencoder_outputs(
    OUTPUT_DIR, TIMESTAMP
)

print("SEAFLOOR CORRELATION VISUALIZATION — 2 LATENT EMBEDDINGS")
print("  LE1 → Biogeochemical        (physicochemical + biological)")
print("  LE2 → Hydrogeomorphological (hydrological  + geomorphological)")
print("=" * 70)
print(f"  Run timestamp      : {resolved_ts}")
print(f"  Correlation matrix : {file_path}")
print(f"  Latent embeddings  : {latent_file_path}")


# =====================================================================
# VARIABLE NAME MAPPING  (internal name → readable English label)
# =====================================================================
column_mapping = {
    # ── Biogeochemical — Physicochemical sub-group ──────────────────
    'Phosphate':            'Phosphate',
    'Iron':                 'Iron',
    'Light_bottom':         'Light at bottom',
    'Nitrate':              'Nitrate',
    'Dissolved_oxygen':     'Dissolved oxygen',
    'pH':                   'pH',
    'Salinity':             'Salinity',
    'Silicate':             'Silicate',
    'Temperature':          'Temperature',
    'P_a_r':                'Photosynthetically active radiation',
    'Organic_matter':       'Organic matter',
    # ── Biogeochemical — Biological sub-group ───────────────────────
    'Chlorophyll':          'Chlorophyll',
    'Phytoplankton':        'Phytoplankton',
    'Utilized_oxygen':      'Utilized oxygen',
    'Primary_productivity': 'Primary productivity',
    # ── Hydrogeomorphological ───────────────────────────────────────
    'Aspect':               'Aspect',
    'Eastness':             'Eastness',
    'River_influence':      'River influence',
    'Currents_velocity':    'Currents velocity',
    'Runoff':               'Runoff',
    'Wind_v':               '10 m v-component of wind',
    'Land_distance':        'Land distance',
    'L_t_m':                'Local terrain mean',
    'Northness':            'Northness',
    'R_d_m_v':              'Relative difference to the mean value',
    'Slope':                'Slope',
    'Stde':                 'Standard deviation of depth',
    'Superficie':           'Surface area',
    'Hardness':             'Hardness',
    'Particle_size':        'Particle size',
    'E_l_d':                'Euphotic layer depth',
    'D_a_c':                'Diffuse attenuation coefficient',
    'R_i_m_b':              'RIMB',   # formerly IBMR — reassigned to hydrogeomorphological
}

# =====================================================================
# CANONICAL VARIABLE LISTS
#   LE1 → BIOGEOCHEMICAL        : physicochemical + biological
#   LE2 → HYDROGEOMORPHOLOGICAL : hydrological + geomorphological
#                                  (R_i_m_b included here)
# =====================================================================
physicochemical_vars = [
    'Phosphate', 'Iron', 'Light_bottom', 'Nitrate', 'Dissolved_oxygen',
    'pH', 'Salinity', 'Silicate', 'Temperature', 'P_a_r', 'Organic_matter'
]
biological_vars = [
    'Chlorophyll', 'Phytoplankton', 'Utilized_oxygen', 'Primary_productivity'
]
biogeochemical_vars = physicochemical_vars + biological_vars           # → LE1

hydrogeomorphological_vars = [
    'Aspect', 'Eastness', 'River_influence', 'Currents_velocity', 'Runoff',
    'Wind_v', 'Land_distance', 'L_t_m', 'Northness', 'R_d_m_v', 'Slope',
    'Stde', 'Superficie', 'Hardness', 'Particle_size', 'E_l_d', 'D_a_c',
    'R_i_m_b'
]                                                                       # → LE2

# =====================================================================
# HELPER FUNCTIONS
# =====================================================================

def strip_prefix(var_name: str) -> str:
    """
    Remove any ALLCAPS_ prefix added by the autoencoder output.

    The autoencoder tags columns with prefixes like FISICO_BIO_ or HIDRO_
    to identify which branch produced them.  This function strips those
    prefixes so we can look the base name up in column_mapping.
    """
    known_prefixes = [
        # New autoencoder naming
        'FISICO_BIO_', 'HIDRO_',
        # Legacy / alternative naming
        'BIOGEOCHEM_', 'HYDROGEO_', 'PHYSBIO_', 'GEO_',
        'FISICO_', 'BIO_', 'HIDRO_', 'PHYSICAL_', 'CHEMICAL_',
        'BIOLOGICAL_', 'HYDRO_', 'GEOMORPH_', 'HYDROGEOMORPH_',
    ]
    for prefix in known_prefixes:
        if var_name.startswith(prefix):
            return var_name[len(prefix):]
    # Generic fallback: strip a leading UPPER_ token
    return re.sub(r'^[A-Z][A-Z0-9]*_', '', var_name)


def clean_variable_name(var_name: str) -> str:
    """Return a human-readable English label for display on the plot."""
    base = strip_prefix(var_name)
    return column_mapping.get(base, base)


def categorize_variable(var_name: str) -> str:
    """
    Classify a variable into one of two latent-embedding categories:
      'Biogeochemical'        → LE1
      'Hydrogeomorphological' → LE2

    Resolution order
    ----------------
    1. Exact prefix match  (fastest, most reliable)
    2. Base-name lookup in the canonical lists
    3. Keyword scan of the lowercased base name
    """
    # 1 ── Fast path via exact prefix ─────────────────────────────────
    if var_name.startswith(('FISICO_BIO_', 'BIOGEOCHEM_', 'PHYSBIO_', 'FISICO_', 'BIO_')):
        return 'Biogeochemical'
    if var_name.startswith(('HIDRO_', 'HYDROGEO_', 'GEO_', 'HYDRO_', 'GEOMORPH_')):
        return 'Hydrogeomorphological'

    # 2 ── Base-name canonical lookup ─────────────────────────────────
    base = strip_prefix(var_name)
    if base in biogeochemical_vars:
        return 'Biogeochemical'
    if base in hydrogeomorphological_vars:
        return 'Hydrogeomorphological'

    # 3 ── Keyword fallback ───────────────────────────────────────────
    b = base.lower()
    biogeo_kw = [
        'phosphate', 'iron', 'nitrate', 'silicate', 'dissolved_oxygen',
        'salinity', 'temperature', 'ph', 'light_bottom', 'p_a_r',
        'organic_matter', 'chlorophyll', 'phytoplankton',
        'utilized_oxygen', 'primary_productivity'
    ]
    hydrogeo_kw = [
        'aspect', 'eastness', 'northness', 'slope', 'stde', 'superficie',
        'l_t_m', 'r_d_m_v', 'hardness', 'particle_size', 'e_l_d', 'd_a_c',
        'currents_velocity', 'wind_v', 'runoff', 'land_distance',
        'river_influence', 'r_i_m_b'
    ]
    if any(k in b for k in biogeo_kw):
        return 'Biogeochemical'
    if any(k in b for k in hydrogeo_kw):
        return 'Hydrogeomorphological'

    return 'Other'


# =====================================================================
# COLOR PALETTE
# =====================================================================
color_map = {
    'Biogeochemical':        '#E74C3C',  # Red  → LE1
    'Hydrogeomorphological': '#3498DB',  # Blue → LE2
    'Other':                 '#95A5A6'   # Grey
}

# =====================================================================
# 1. LOAD CORRELATION MATRIX
#
# The autoencoder writes the matrix with latent embeddings as ROWS and
# original variables as COLUMNS.  We transpose so each row = 1 variable.
# =====================================================================
df_raw = pd.read_csv(file_path, index_col=0)

# Detect orientation: if the index contains known LE names, rows = LEs
# and we need to transpose; otherwise it is already variable-per-row.
index_has_le = df_raw.index.str.contains(
    r'LE[12]|FisicoBio|Hidro', case=False, regex=True
).any()

if index_has_le:
    # Standard autoencoder output: rows = LEs, cols = variables → transpose
    df = df_raw.T.reset_index()
    df.rename(columns={'index': 'Variables'}, inplace=True)
else:
    # Already variable-per-row
    df = df_raw.reset_index()
    df.rename(columns={'index': 'Variables'}, inplace=True)

print(f"\nData loaded   : {df.shape[0]} variables × {df.shape[1]-1} latent embedding(s)")
print(f"Columns found : {df.columns.tolist()}")

# =====================================================================
# 2. CATEGORIZE AND RENAME
# =====================================================================
df['Category']        = df['Variables'].apply(categorize_variable)
df['Variables_clean'] = df['Variables'].apply(clean_variable_name)

print("\nCategorization results:")
label_map = {
    'Biogeochemical':        'Biogeochemical (LE1)',
    'Hydrogeomorphological': 'Hydrogeomorphological (LE2)',
    'Other':                 'Other'
}
for cat, count in df['Category'].value_counts().items():
    print(f"  {label_map.get(cat, cat)}: {count} variables")

# Identify LE column names dynamically (exclude metadata columns)
meta_cols = {'Variables', 'Category', 'Variables_clean'}
le_cols   = [c for c in df.columns if c not in meta_cols]

if len(le_cols) < 2:
    raise ValueError(
        f"Expected at least 2 latent-embedding columns but found {len(le_cols)}: {le_cols}.\n"
        f"Check that the correlation file has LE1 and LE2 as rows (or columns)."
    )

le1_col, le2_col = le_cols[0], le_cols[1]
print(f"\nLatent columns → LE1: '{le1_col}'  |  LE2: '{le2_col}'")

# =====================================================================
# 3. FIGURE: TWO BAR-PLOT PANELS
# =====================================================================
fig, axes = plt.subplots(2, 1, figsize=(22, 22))
fig.patch.set_facecolor('white')

# ── Panel A: LE1 — BIOGEOCHEMICAL ────────────────────────────────────
df_le1 = df.sort_values(le1_col, ascending=False).reset_index(drop=True)
axes[0].bar(
    range(len(df_le1)), df_le1[le1_col].values,
    color=[color_map.get(c, color_map['Other']) for c in df_le1['Category']],
    alpha=0.85, edgecolor='white', linewidth=0.5
)
axes[0].set_title(
    'LE1 — Biogeochemical\n(Physicochemical + Biological variables)',
    fontsize=20, fontweight='bold', pad=40
)
axes[0].set_ylabel('Signed Spearman correlation', fontsize=15, fontweight='bold')
axes[0].set_xticks(range(len(df_le1)))
axes[0].set_xticklabels(df_le1['Variables_clean'], rotation=90, ha='right', fontsize=9)
axes[0].axhline(0, color='black', linewidth=0.8)
axes[0].grid(axis='y', linestyle='--', alpha=0.55)

legend_elements = [
    Patch(facecolor=color_map['Biogeochemical'],        label='Biogeochemical (LE1)',        alpha=0.85),
    Patch(facecolor=color_map['Hydrogeomorphological'], label='Hydrogeomorphological (LE2)', alpha=0.85),
    Patch(facecolor=color_map['Other'],                 label='Other',                       alpha=0.85),
]
axes[0].legend(handles=legend_elements, loc='upper right', fontsize=12,
               title='VARIABLE CATEGORY', title_fontsize=13, framealpha=0.9)

# ── Panel B: LE2 — HYDROGEOMORPHOLOGICAL ─────────────────────────────
df_le2 = df.sort_values(le2_col, ascending=False).reset_index(drop=True)
axes[1].bar(
    range(len(df_le2)), df_le2[le2_col].values,
    color=[color_map.get(c, color_map['Other']) for c in df_le2['Category']],
    alpha=0.85, edgecolor='white', linewidth=0.5
)
axes[1].set_title(
    'LE2 — Hydrogeomorphological\n(Hydrological + Geomorphological variables, incl. R_i_m_b)',
    fontsize=20, fontweight='bold', pad=40
)
axes[1].set_ylabel('Signed Spearman correlation', fontsize=15, fontweight='bold')
axes[1].set_xlabel('Variables', fontsize=15, fontweight='bold')
axes[1].set_xticks(range(len(df_le2)))
axes[1].set_xticklabels(df_le2['Variables_clean'], rotation=90, ha='right', fontsize=9)
axes[1].axhline(0, color='black', linewidth=0.8)
axes[1].grid(axis='y', linestyle='--', alpha=0.55)

plt.subplots_adjust(top=0.95, bottom=0.08, left=0.07, right=0.97, hspace=0.88)

# =====================================================================
# 4. SAVE FIGURE
# =====================================================================
output_fig = os.path.join(OUTPUT_DIR, f"CORR_BARPLOT_2LE_{resolved_ts}.png")
plt.savefig(output_fig, dpi=300, bbox_inches='tight', facecolor='white')
print(f"\nFigure saved: {output_fig}")
plt.show()

# =====================================================================
# 5. SPECIALIZATION ANALYSIS
# =====================================================================
print("\n" + "=" * 70)
print("SPECIALIZATION ANALYSIS")
print("  Metric: mean absolute signed-correlation per category")
print("=" * 70)

# LE1 — Biogeochemical
le1_bio   = df[df['Category'] == 'Biogeochemical'][le1_col].abs().mean()
le1_geo   = df[df['Category'] == 'Hydrogeomorphological'][le1_col].abs().mean()
le1_ratio  = le1_bio / le1_geo if le1_geo > 0 else float('inf')
le1_purity = max(0.0, 1 - (le1_geo / (le1_bio + 1e-6)))

print(f"\nLE1 (Biogeochemical):")
print(f"  Mean |corr| with Biogeochemical vars      : {le1_bio:.4f}")
print(f"  Mean |corr| with Hydrogeomorphol. vars    : {le1_geo:.4f}")
print(f"  Specialization ratio  (target > 1.5)      : {le1_ratio:.2f}")
print(f"  Purity score          (target > 0.35)     : {le1_purity:.4f}  "
      f"{'[OK]' if le1_purity > 0.35 else '[REVIEW]'}")

# LE2 — Hydrogeomorphological
le2_geo   = df[df['Category'] == 'Hydrogeomorphological'][le2_col].abs().mean()
le2_bio   = df[df['Category'] == 'Biogeochemical'][le2_col].abs().mean()
le2_ratio  = le2_geo / le2_bio if le2_bio > 0 else float('inf')
le2_purity = max(0.0, 1 - (le2_bio / (le2_geo + 1e-6)))

print(f"\nLE2 (Hydrogeomorphological):")
print(f"  Mean |corr| with Hydrogeomorphol. vars    : {le2_geo:.4f}")
print(f"  Mean |corr| with Biogeochemical vars      : {le2_bio:.4f}")
print(f"  Specialization ratio  (target > 1.5)      : {le2_ratio:.2f}")
print(f"  Purity score          (target > 0.30)     : {le2_purity:.4f}  "
      f"{'[OK]' if le2_purity > 0.30 else '[CRITICAL]'}")

# R_i_m_b spot-check
rimb = df[df['Variables'].str.contains('R_i_m_b', case=False, na=False)]
if not rimb.empty:
    print(f"\nR_i_m_b (RIMB) — Hydrogeomorphological spot-check:")
    print(f"  Category : {rimb['Category'].values[0]}")
    print(f"  LE1 corr : {rimb[le1_col].values[0]:.4f}")
    print(f"  LE2 corr : {rimb[le2_col].values[0]:.4f}")

# Combined summary
combined = (le1_ratio + le2_ratio) / 2
status   = ("EXCELLENT"    if combined > 3.0 else
            "GOOD"         if combined > 2.0 else
            "PROGRESS"     if combined > 1.2 else
            "INSUFFICIENT")

print(f"\nCombined specialization score : {combined:.2f}")
print(f"Overall status                : {status}")
print("=" * 70)
```

# 3. Orthogonality Between Latent Features

```python
import os
import re
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import QuantileTransformer
from datetime import datetime

# =====================================================================
# FILE DISCOVERY — reads the LATEST latent-embedding output from the
# autoencoder.
#
# The autoencoder writes:
#   LE_DUAL_OPTIMIZED_{timestamp}.csv          ← current naming
#   LE_DUAL_OPTIMIZED_R_i_m_b_{timestamp}.csv  ← legacy naming
#
# Both live in:  data/outputs/
#
# Set TIMESTAMP to a specific string (e.g. "20260223_161913") to pin
# a run, or leave as None for automatic discovery of the latest file.
# =====================================================================

OUTPUT_DIR = "data/outputs"
TIMESTAMP  = None   # ← set to e.g. "20260223_161913" to pin a specific run


def find_latest_latent_file(output_dir: str, timestamp: str | None):
    """
    Returns (latent_path, resolved_timestamp).

    File-name variants accepted
    ---------------------------
      LE_DUAL_OPTIMIZED_{ts}.csv           current autoencoder output
      LE_DUAL_OPTIMIZED_R_i_m_b_{ts}.csv  legacy name (old runs)
    """

    def _find_latent(ts: str):
        patterns = [
            f"LE_DUAL_OPTIMIZED_{ts}.csv",
            f"LE_DUAL_OPTIMIZED_R_i_m_b_{ts}.csv",
        ]
        for p in patterns:
            fp = os.path.join(output_dir, p)
            if os.path.exists(fp):
                return fp
        return None

    # ── Case 1: explicit timestamp ────────────────────────────────────
    if timestamp is not None:
        fp = _find_latent(timestamp)
        if fp is None:
            raise FileNotFoundError(
                f"Latent-embedding file not found for timestamp '{timestamp}' "
                f"in '{output_dir}'.\n"
                f"Expected one of:\n"
                f"  LE_DUAL_OPTIMIZED_{timestamp}.csv\n"
                f"  LE_DUAL_OPTIMIZED_R_i_m_b_{timestamp}.csv"
            )
        return fp, timestamp

    # ── Case 2: auto-discover the most recent file ────────────────────
    le_glob = glob.glob(os.path.join(output_dir, "LE_DUAL_OPTIMIZED_*.csv"))
    if not le_glob:
        raise FileNotFoundError(
            f"No 'LE_DUAL_OPTIMIZED_*.csv' files found in '{output_dir}'.\n"
            f"Run the autoencoder first to generate outputs."
        )

    ts_pattern = re.compile(r"LE_DUAL_OPTIMIZED_(?:R_i_m_b_)?(\d{8}_\d{6})\.csv$")
    candidates = []
    for fp in le_glob:
        m = ts_pattern.search(os.path.basename(fp))
        if m:
            candidates.append((m.group(1), fp))

    if not candidates:
        raise FileNotFoundError(
            f"Found LE_DUAL_OPTIMIZED_*.csv files but could not parse timestamps.\n"
            f"Files found:\n" + "\n".join(le_glob)
        )

    # Pick the most recent
    candidates.sort(key=lambda t: t[0], reverse=True)
    resolved_ts, resolved_fp = candidates[0]
    return resolved_fp, resolved_ts


# ── Resolve path ──────────────────────────────────────────────────────
latent_file_path, resolved_ts = find_latest_latent_file(OUTPUT_DIR, TIMESTAMP)

print("LE1 vs LE2 — SCATTER PLOT & ORTHOGONALITY ANALYSIS")
print("  LE1 → Biogeochemical        (physicochemical + biological)")
print("  LE2 → Hydrogeomorphological (hydrological  + geomorphological)")
print("=" * 70)
print(f"  Run timestamp : {resolved_ts}")
print(f"  Input file    : {latent_file_path}")


# =====================================================================
# SETUP RPY2 AND MINERVA FOR MIC
# =====================================================================
try:
    import rpy2
    import rpy2.robjects as ro
    from rpy2.robjects import pandas2ri
    from rpy2.robjects.packages import importr

    pandas2ri.activate()
    ro.r('options(repos = c(CRAN = "https://cran.r-project.org"))')
    ro.r('options(warn = -1)')
    utils = importr('utils')

    try:
        if not ro.r('any(grepl("minerva", installed.packages()))')[0]:
            print("Installing minerva...")
            ro.r('install.packages("minerva", dependencies=TRUE, quiet=TRUE)')
        minerva = importr('minerva')
        print("rpy2 + Minerva available — MIC will be computed")
        RPY2_AVAILABLE = True
    except Exception as e:
        print(f"Minerva error: {e}")
        RPY2_AVAILABLE = False
except ImportError:
    print("rpy2 not available — Pearson correlation only")
    RPY2_AVAILABLE = False


# =====================================================================
# MIC HELPER FUNCTIONS
# =====================================================================

def setup_signed_realistic_mic():
    if not RPY2_AVAILABLE:
        return False
    try:
        r_mic_code = """
        suppressWarnings(library(minerva, quietly = TRUE))
        calculate_signed_realistic_mic <- function(original_df, latent_df) {
            orig_cols   <- names(original_df)
            latent_cols <- names(latent_df)
            mic_matrix  <- matrix(0.0, nrow=length(latent_cols), ncol=length(orig_cols))
            colnames(mic_matrix) <- orig_cols
            rownames(mic_matrix) <- latent_cols
            for (i in seq_along(latent_cols)) {
                y <- latent_df[[latent_cols[i]]]
                for (j in seq_along(orig_cols)) {
                    x <- original_df[[orig_cols[j]]]
                    if (any(is.na(x)) || any(is.infinite(x)) || sd(x)==0 ||
                        any(is.na(y)) || any(is.infinite(y)) || sd(y)==0) { next }
                    suppressWarnings({
                        tryCatch({
                            mine_result <- minerva::mine(x, y, alpha=0.6, C=15)
                            mic_value   <- mine_result$MIC
                            correlation <- cor(x, y, method="spearman")
                            signed_mic  <- ifelse(correlation >= 0, mic_value, -mic_value)
                            mic_matrix[i, j] <- signed_mic
                        }, error=function(e){})
                    })
                }
            }
            return(as.data.frame(mic_matrix))
        }
        """
        with open("/tmp/signed_realistic_mic.R", "w") as f:
            f.write(r_mic_code)
        ro.r.source("/tmp/signed_realistic_mic.R")
        return True
    except Exception as e:
        print(f"Error configuring signed MIC: {e}")
        return False


def calculate_mic_between_variables(x: np.ndarray, y: np.ndarray):
    """Return signed MIC between two 1-D arrays, or None if unavailable."""
    if not RPY2_AVAILABLE:
        return None
    try:
        df_x = pd.DataFrame({'var_x': x.copy()})
        df_y = pd.DataFrame({'var_y': y.copy()})

        for df in [df_x, df_y]:
            for col in df.columns:
                data = df[col].values
                if np.var(data) > 1e-12:
                    qt = QuantileTransformer(
                        n_quantiles=min(300, len(data) // 2),
                        output_distribution='uniform', random_state=42
                    )
                    df[col] = qt.fit_transform(data.reshape(-1, 1)).flatten()

        df_x = df_x.fillna(0.5).replace([np.inf, -np.inf], 0.5)
        df_y = df_y.fillna(0.5).replace([np.inf, -np.inf], 0.5)

        mic_result = ro.r['calculate_signed_realistic_mic'](df_x, df_y)
        mic_matrix = pandas2ri.rpy2py(mic_result)
        return float(mic_matrix.iloc[0, 0])
    except Exception as e:
        print(f"Error calculating MIC: {e}")
        return None


# =====================================================================
# LOAD DATA & IDENTIFY LE COLUMNS
#
# The autoencoder names the latent columns:
#   LE1_FisicoBio_OPTIMIZED   (current)   or   LE1_Biogeochem_*  (legacy)
#   LE2_Hidro_OPTIMIZED       (current)   or   LE2_Hydro_*       (legacy)
# =====================================================================
mic_available = setup_signed_realistic_mic()

df = pd.read_csv(latent_file_path)
print(f"\nDataset loaded: {df.shape[0]} rows × {df.shape[1]} columns")
print(f"Columns: {df.columns.tolist()}")

# Auto-detect LE1 column — prefer specific names, then fallback to any LE1
le1_candidates = (
    [c for c in df.columns if re.search(r'LE1.*Fisico|LE1.*Bio', c, re.I)]
    or [c for c in df.columns if c.upper().startswith('LE1')]
)
# Auto-detect LE2 column
le2_candidates = (
    [c for c in df.columns if re.search(r'LE2.*Hidro|LE2.*Hydro', c, re.I)]
    or [c for c in df.columns if c.upper().startswith('LE2')]
)

if not le1_candidates or not le2_candidates:
    available_le = [c for c in df.columns if 'LE' in c.upper()]
    raise ValueError(
        f"Could not find LE1 / LE2 columns.\n"
        f"LE-related columns found: {available_le}\n"
        f"All columns: {df.columns.tolist()}"
    )

le1_col = le1_candidates[0]
le2_col = le2_candidates[0]
print(f"\n  LE1 column : {le1_col}")
print(f"  LE2 column : {le2_col}")

x = df[le1_col].dropna()
y = df[le2_col].dropna()

# Align indices
common_idx = x.index.intersection(y.index)
x, y = x.loc[common_idx], y.loc[common_idx]

# =====================================================================
# STATISTICS
# =====================================================================
corr_coef, p_value = pearsonr(x, y)

mic_value = None
if mic_available:
    print("\nCalculating MIC between LE1 and LE2...")
    mic_value = calculate_mic_between_variables(x.values, y.values)
    if mic_value is not None:
        print(f"  MIC = {mic_value:.4f}")
    else:
        print("  MIC calculation failed — Pearson only")

# Regression
reg = LinearRegression()
reg.fit(x.values.reshape(-1, 1), y.values)
r_squared = reg.score(x.values.reshape(-1, 1), y.values)
x_trend   = np.linspace(x.min(), x.max(), 200)
y_trend   = reg.predict(x_trend.reshape(-1, 1))

# =====================================================================
# FIGURE
# =====================================================================
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['font.size']   = 10
plt.rcParams['axes.linewidth'] = 1.2

fig, ax = plt.subplots(figsize=(8, 6))
fig.patch.set_facecolor('white')

# Scatter
ax.scatter(x, y, alpha=0.50, s=30, color='#2E86AB',
           edgecolors='none', label='Observations', zorder=3)

# Regression line
ax.plot(x_trend, y_trend, color='#A23B72', linewidth=2.5, zorder=5,
        label=f'Linear fit  (R² = {r_squared:.3f})')

# Axis labels
ax.set_xlabel(
    'LE1 — Biogeochemical\n(Physicochemical + Biological variables)',
    fontsize=11
)
ax.set_ylabel(
    'LE2 — Hydrogeomorphological\n(Hydrological + Geomorphological variables)',
    fontsize=11
)
ax.set_title(
    'Orthogonality Between Latent Embeddings\n'
    'Dual Autoencoder — Biogeochemical vs Hydrogeomorphological',
    fontsize=12, fontweight='bold', pad=15
)

# Grid
ax.grid(True, alpha=0.25, linestyle='--', linewidth=0.5)

# Legend
ax.legend(loc='upper left', frameon=True, framealpha=0.95,
          edgecolor='black', fontsize=9)

# Statistics text box
stats_lines = [
    f'n = {len(x)}',
    f'Pearson r = {corr_coef:.4f}',
    f'p-value   = {p_value:.4f}',
]
if mic_value is not None:
    stats_lines.append(f'MIC       = {mic_value:.4f}')
else:
    stats_lines.append('MIC       = N/A (rpy2 unavailable)')

ax.text(0.97, 0.03, '\n'.join(stats_lines),
        transform=ax.transAxes, fontsize=9,
        verticalalignment='bottom', horizontalalignment='right',
        bbox=dict(boxstyle='round', facecolor='white', alpha=0.92,
                  edgecolor='black', linewidth=1))

# Spines
for spine in ax.spines.values():
    spine.set_visible(True)
    spine.set_color('black')
    spine.set_linewidth(1.2)

plt.tight_layout()

# =====================================================================
# SAVE FIGURE  →  data/outputs/
# =====================================================================
run_ts   = datetime.now().strftime("%Y%m%d_%H%M%S")
out_name = f"Figure_LE1_LE2_Orthogonality_{run_ts}.png"
out_path = os.path.join(OUTPUT_DIR, out_name)

fig.savefig(out_path, dpi=300, bbox_inches='tight',
            facecolor='white', edgecolor='none', pad_inches=0.1)
print(f"\nFigure saved: {out_path}  (300 DPI)")
plt.show()

# =====================================================================
# STATISTICAL REPORT
# =====================================================================
print("\n" + "=" * 70)
print("DESCRIPTIVE STATISTICS")
print("=" * 70)
for col, data in [(le1_col, x), (le2_col, y)]:
    tag = "LE1 — Biogeochemical" if 'LE1' in col else "LE2 — Hydrogeomorphological"
    print(f"\n{tag}:")
    print(f"  Mean   : {data.mean():.4f}")
    print(f"  Std    : {data.std():.4f}")
    print(f"  Min    : {data.min():.4f}")
    print(f"  Max    : {data.max():.4f}")
    print(f"  Median : {data.median():.4f}")

print("\n" + "=" * 70)
print("CORRELATION ANALYSIS")
print("=" * 70)

def interpret_strength(value):
    v = abs(value)
    return "Strong" if v >= 0.7 else "Moderate" if v >= 0.3 else "Weak"

pearson_interp = interpret_strength(corr_coef)
print(f"\nPearson r  = {corr_coef:.4f}  →  {pearson_interp} correlation")
print(f"p-value    = {p_value:.4f}  →  {'Significant' if p_value < 0.05 else 'Not significant'} (α=0.05)")
print(f"R²         = {r_squared:.4f}")

if mic_value is not None:
    mic_interp = interpret_strength(mic_value)
    ratio_mic_pearson = abs(mic_value) / abs(corr_coef) if abs(corr_coef) > 1e-6 else 0
    print(f"\nMIC        = {mic_value:.4f}  →  {mic_interp} dependence")
    print(f"MIC/|r|    = {ratio_mic_pearson:.2f}  →  "
          f"{'Non-linear relationships detected' if ratio_mic_pearson > 1.2 else 'Predominantly linear'}")

print("\n" + "=" * 70)
print("ORTHOGONALITY EVALUATION")
print("=" * 70)

def orthogonality_label(r):
    r = abs(r)
    if r < 0.10: return "Excellent"
    if r < 0.30: return "Good"
    if r < 0.50: return "Moderate"
    return "Limited"

print(f"\nOrthogonality (Pearson) : {orthogonality_label(corr_coef)}")
if mic_value is not None:
    print(f"Independence (MIC)      : {orthogonality_label(mic_value)}")

separation = ('Well' if abs(corr_coef) < 0.3 else
              'Moderately' if abs(corr_coef) < 0.5 else 'Poorly')
print(f"LE1–LE2 separation      : {separation} separated")

print("\n" + "=" * 70)
print("VARIANCE ANALYSIS")
print("=" * 70)
var1 = x.var()
var2 = y.var()
print(f"\nVariance LE1 (Biogeochemical)        : {var1:.4f}")
print(f"Variance LE2 (Hydrogeomorphological) : {var2:.4f}")
print(f"Variance ratio (LE1/LE2)             : {var1/var2:.2f}  "
      f"{'[Balanced]' if 0.5 < var1/var2 < 2.0 else '[Imbalanced]'}")

print("\n" + "=" * 70)
print("FINAL SUMMARY")
print("=" * 70)
print(f"  LE1–LE2 Pearson r      : {corr_coef:.4f}  ({pearson_interp})")
if mic_value is not None:
    print(f"  LE1–LE2 MIC            : {mic_value:.4f}  ({mic_interp})")
print(f"  Orthogonality status   : {orthogonality_label(corr_coef)}")
print(f"  Effective separation   : {'YES' if abs(corr_coef) < 0.3 else 'MODERATE' if abs(corr_coef) < 0.5 else 'NO'}")
print(f"  Balanced variances     : {'YES' if 0.5 < var1/var2 < 2.0 else 'NO'}")
if mic_value is not None and abs(mic_value) > abs(corr_coef) * 1.2:
    print("  Note: MIC > |Pearson| — possible non-linear LE1–LE2 relationships")
print("=" * 70)
```
